# Stack Overflow Developer Survey 2022

Dataset: Stack Overflow survey data 2022.

> Project goal is to find out:

Who did the survey?

Which country do the survey participants come from?

How long have the survey participants coded?

### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Importing dataset

```python
survey_2022 = pd.read_csv('stack-overflow-developer-survey-2022/survey_results_public.csv')
survey_2022.head()
```

```python
survey_2022.shape
```

```python
survey_2022.describe()
```

### Data Analysis

```python
columns = ['ResponseId' , 'MainBranch'  , 'YearsCode' , 'Country']

survey_2022_ = survey_2022[columns]
```

```python
survey_2022_.info()
```

```python
survey_2022_["YearsCode"].fillna(survey_2022_["YearsCode"].mean)
survey_2022_["Country"].fillna("Other")
```

```python
survey_2022_.head()
```

### Data Visualization

1. Who did the survey?

```python
ax = sns.countplot(data = survey_2022_, x = 'MainBranch' );
plt.xticks(rotation = 90)
plt.title('Who did the survey?')
plt.xlabel('Survey participants')
plt.ylabel('Number of participants')
plt.savefig('fig1.png', bbox_inches='tight')
plt.show()
```

2. Which country do the survey participants come from?

```python
plt.figure(figsize=(30,10))
ax = sns.countplot(data = survey_2022_, x = 'Country' );
plt.xticks(rotation = 90)
plt.title('Which country do the survey participants come from')
plt.xlabel('Country')
plt.ylabel('Number of participants')
plt.savefig('fig2.png', bbox_inches='tight')
plt.show()
```

3. How long have the survey participants coded?

```python
plt.figure(figsize=(30,10))
ax = sns.countplot(data = survey_2022_, x = 'YearsCode' );
plt.xticks(rotation = 90)
plt.title('How long have the survey participants coded')
plt.xlabel('Years Code')
plt.ylabel('Number of participants')
plt.savefig('fig3.png', bbox_inches='tight')
plt.show()
```

### Conclusions:

These are some observation for the mentioned questions:

- People who is a developer by profession do the survey the most
- People who are from US do the survey the most
- People who have coded for 5 and 10 years do the survey the most
